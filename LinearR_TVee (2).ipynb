FEATURE ENGINEERING

```python
# Data Splitting

from sklearn.model_selection import train_test_split

# X is our independent variable (feature)
X = df2.drop('Salary', axis=1)

# y is our dependent variable (label)
y = df2['Salary']

# Split in 80:20
# Train set ===> X_train, y_train
X_train, X_test, y_train, y_test = train_test_split(df2.drop(columns=['Salary']), y, test_size=0.2, random_state=42)
# Test set ===> X_test, y_test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_test.head()
```

WITHOUT SCALING

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_train, y_train)
```

```python
linreg.score(X_test, y_test)
```

```python
linreg.predict([[45, 0, 3, 15]])
```

```python
linreg.predict([[28, 1, 2, 3]])
```

WITH SCALING

```python
# MinMaxScaler - used to scale to 0 & 1

from sklearn.preprocessing import MinMaxScaler
```

```python
# Instantiate the scaler object

scaler = MinMaxScaler()
```

```python
# Fit on train
scaler.fit(X_train)

# Transform on train and test
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
linreg = LinearRegression()

linreg.fit(X_train_scaled, y_train)
```

```python
linreg.score(X_test_scaled, y_test)
```

```python
unscaled = 0.9097525936677504
scaled = 0.9097525936677505

scaled > unscaled
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Fetch the predictions on the test set

preds = linreg.predict(X_test)
```

```python
print(mean_squared_error(y_test, preds))
```

```python
#ROOT MEAN SQUARE ERROR
# RMSE

print(np.sqrt(mean_squared_error(y_test, preds)))
```

```python
#MEAN ABSOLUTE ERROR

print(mean_absolute_error(y_test, preds))
```
